#1 What is the average revenue generated by each branch?

```python
average_revenue = df.groupby('branch')['revenue'].mean().reset_index(name='average_revenue')
```

```python
plt.bar(average_revenue['branch'], average_revenue['average_revenue'])
plt.title('Average Revenue by Branch')
plt.xlabel('Branch')
plt.ylabel('Average Revenue')
plt.show()
```

#2 Which product line generates the highest revenue?

```python
product_revenue = df.groupby('product_line')['revenue'].sum().reset_index()
```

```python
product_revenue = product_revenue.sort_values(by='revenue', ascending=False)
```

```python
plt.figure(figsize=(8, 4))
plt.barh(product_revenue['product_line'], product_revenue['revenue'], color='skyblue')
plt.title('Revenue by Product Line')
plt.xlabel('Total Revenue')
plt.ylabel('Product Line')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest revenue at the top
plt.show()
```

#3 Are there any differences in spending habits between male and female customers?

```python
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['gender_customer'] == 'Male']['revenue'], df[df['gender_customer'] == 'Female']['revenue']], labels=['Male', 'Female'])
plt.title('Spending Habits by Gender')
plt.xlabel('Gender')
plt.ylabel('Revenue')
plt.show()
```

#4 What is the most common payment method used by customers?

```python
payment_counts = df['payment_method'].value_counts()
```

```python
most_common_payment = payment_counts.idxmax()
```

```python
plt.bar(payment_counts.index, payment_counts.values, color='blue')
plt.title('Most Common Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Frequency')
plt.show()
```
